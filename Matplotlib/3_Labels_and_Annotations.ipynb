# **Matplotlib Labels and Annotations Tutorial**


## **1. Introduction to Matplotlib Labels**

### What Are Labels in Data Visualization?
Labels in data visualization are textual descriptions used to:
- Identify data points or axes.
- Provide context through titles, subtitles, and legends.
- Highlight specific features or areas of interest using annotations.

### Why Are Labels Important?
- **Enhance interpretability:** Labels make it easier for viewers to understand the data.
- **Communicate effectively:** Provide necessary information without cluttering the plot.
- **Professional presentation:** Well-labeled plots are visually appealing and easier to read.

### Types of Labels in Matplotlib
1. **Axis Labels:** Describe the data represented on each axis (`xlabel`, `ylabel`).
2. **Titles and Subtitles:** Provide the main context for the chart.
3. **Legends:** Explain what different elements (e.g., lines, markers) represent.
4. **Annotations:** Highlight specific data points or regions with custom text.


## **2. Basic Examples of Labels**

### **2.1 Adding Basic Axis Labels and Titles**

```python

import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a simple line plot
plt.plot(x, y, label="Sine Wave")
plt.title("Basic Plot Example")  # Add a title
plt.xlabel("X-axis (radians)")   # Add x-axis label
plt.ylabel("Y-axis (amplitude)") # Add y-axis label
plt.legend()  # Add legend
plt.show()
```

### **2.2 Customizing Titles and Labels**

```python

# Customizing plot labels
plt.plot(x, y, label="Sine Wave", color="orange")
plt.title("Customized Title", fontsize=18, color='blue', fontweight='bold', loc='center')
plt.xlabel("X-axis (radians)", fontsize=12, color='green', fontweight='light')
plt.ylabel("Y-axis (amplitude)", fontsize=12, color='purple', fontstyle='italic')
plt.legend(loc="upper right")
plt.show()
```

## **3. Legend Creation and Customization**

### **3.1 Adding and Positioning Legends**

```python

# Plot with a customized legend
plt.plot(x, y, label="Sine Wave", color="red")
plt.title("Legend Customization Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc='lower right', fontsize=10, frameon=True, shadow=True, title="Legend Title")
plt.show()
```

## **4. Adding Annotations**

### **4.1 Simple Text Annotations**

```python

# Annotate a specific point
plt.plot(x, y, label="Sine Wave", color="orange")
plt.title("Annotation Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Highlight a peak in the sine wave
max_x = np.pi / 2
max_y = np.sin(max_x)
plt.scatter(max_x, max_y, color="red")  # Highlight point
plt.annotate(
    "Max Point",
    xy=(max_x, max_y),
    xytext=(2, 0.8),
    arrowprops=dict(facecolor='black', arrowstyle="->"),
    fontsize=10
)
plt.legend()
plt.show()
```

## **5. Working with Multiple Subplots**

### **5.1 Subplots with Different Label Styles**

```python

# Create subplots with unique titles and labels
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# First subplot
ax[0, 0].plot(x, y, color='blue')
ax[0, 0].set_title("Plot 1")
ax[0, 0].set_xlabel("X-axis 1")
ax[0, 0].set_ylabel("Y-axis 1")

# Second subplot
ax[0, 1].bar([1, 2, 3], [3, 4, 5], color='green')
ax[0, 1].set_title("Plot 2")
ax[0, 1].set_xlabel("X-axis 2")

# Third subplot
ax[1, 0].scatter(x, y, color='purple')
ax[1, 0].set_title("Plot 3")
ax[1, 0].set_xlabel("X-axis 3")
ax[1, 0].set_ylabel("Y-axis 3")

# Fourth subplot
ax[1, 1].plot(x, y**2, color='red')
ax[1, 1].set_title("Plot 4")
ax[1, 1].set_xlabel("X-axis 4")
ax[1, 1].set_ylabel("Y-axis 4")

# Adjust spacing
plt.tight_layout()
plt.show()
```

## **6. Practical Tips for Readability**


- Use concise and clear labels.
- Avoid overlapping labels by using rotation or alignment.
- Use annotations sparingly to prevent clutter.
- Choose font sizes and colors that contrast well with the plot background.


### Example: Rotated Labels for Bar Charts

```python

# Bar chart with rotated labels
categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 7]

plt.bar(categories, values, color="skyblue")
plt.title("Bar Chart with Rotated Labels")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(rotation=45, ha='right')
plt.show()
```
