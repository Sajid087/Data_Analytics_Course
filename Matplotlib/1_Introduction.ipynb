# Introduction to Matplotlib

## Overview
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It's one of the most widely used data visualization libraries and provides a MATLAB-like interface for creating plots.

Key features:
- Creation of publication-quality figures
- Support for various plot types
- High degree of customization
- Integration with NumPy and Pandas


````python
# Import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Enable inline plotting in Jupyter
%matplotlib inline
````

## Basic Usage

Let's create a simple line plot to demonstrate the basic usage of Matplotlib.

````python
# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Simple Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

## Customization

Matplotlib offers extensive customization options. Here are some common customizations:

````python
# Create a more customized plot
plt.figure(figsize=(12, 6))

# Plot multiple lines with different styles
plt.plot(x, np.sin(x), 'r--', label='sin(x)', linewidth=2)
plt.plot(x, np.cos(x), 'b-', label='cos(x)', linewidth=2)

# Customize the plot
plt.title('Sine and Cosine Waves', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)

# Add axis limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

plt.show()
````

### Common Customization Options:

1. **Colors**: Use color names ('red', 'blue') or codes ('r', 'b')
2. **Line Styles**: '-' (solid), '--' (dashed), ':' (dotted), '-.' (dash-dot)
3. **Markers**: 'o' (circle), 's' (square), '^' (triangle)
4. **Figure Size**: `figsize=(width, height)` in inches
5. **Font Sizes**: `fontsize` parameter for text elements
6. **Grid**: `grid()` function with various styling options

````python
# Example with multiple subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# First subplot
ax1.plot(x, np.sin(x), 'g-', marker='o', markersize=4)
ax1.set_title('Sine Wave with Markers')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.grid(True)

# Second subplot
ax2.plot(x, np.cos(x), 'r-', linewidth=2)
ax2.set_title('Cosine Wave with Custom Style')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.grid(True)

plt.tight_layout()
plt.show()
````
