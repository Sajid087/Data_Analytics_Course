# Pie Charts in Matplotlib: A Comprehensive Tutorial

Pie charts are a popular data visualization tool used to display proportions or percentages. In this tutorial, we cover:

1. **Introduction to Pie Charts**

4. **Common Customization Options in Matplotlib**

## 1. Introduction to Pie Charts

A pie chart is a circular chart divided into wedge-like sectors, illustrating numerical proportions. The size of each sector represents the proportion of a specific category in the dataset.

### Advantages of Pie Charts
- Easy to understand.
- Effective for showing relative proportions.
- Useful for small datasets.

### Disadvantages of Pie Charts
- Difficult to compare similar-sized wedges.
- Not suitable for large datasets or detailed comparisons.

## 2. Basic Anatomy of a Pie Chart

Key components of a pie chart:
- **Wedges**: Represent individual categories.
- **Labels**: Identify the categories.
- **Colors**: Differentiate the wedges visually.
- **Percentages**: Show the proportion of each wedge.


```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Sample data for a basic pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'lightblue', 'lightgreen', 'pink']

# Create a basic pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Basic Pie Chart')
plt.show()
```

## 3. When to Use (and When Not to Use) Pie Charts

### When to Use:
- Highlighting proportions in a small dataset.
- Comparing parts of a whole.
- Communicating data to a general audience.

### When Not to Use:
- Comparing numerous categories.
- Displaying detailed or precise data.
- Visualizing changes over time.

## 4. Common Customization Options in Matplotlib

### Example: Customizing Colors, Labels, and Explode Effects

```python
# Customizing colors, labels, and explode effects
explode = (0, 0.1, 0, 0)  # Only explode the 2nd slice
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140)
plt.title('Customized Pie Chart')
plt.show()
```

### Example: Adding Legends and Shadows

```python
# Adding a shadow and a legend
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Pie Chart with Shadow and Legend')
plt.legend(title='Categories', loc='best')
plt.show()
```

## 5. Advanced Examples and Styling Techniques

### Example: Creating Donut Charts

```python
# Donut chart (pie chart with a hole in the center)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.3})
plt.title('Donut Chart')
plt.show()
```

### Example: Multiple Pie Charts in Subplots

```python
# Multiple pie charts in subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[0].set_title('Pie Chart 1')

axs[1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=180, shadow=True)
axs[1].set_title('Pie Chart 2 with Shadow')

plt.show()
```

## Best Practices and Troubleshooting Tips

- Use pie charts sparingly for clear comparisons.
- Avoid using too many categories to prevent clutter.
- Use contrasting colors for better visibility.
- Add labels or legends to improve readability.
